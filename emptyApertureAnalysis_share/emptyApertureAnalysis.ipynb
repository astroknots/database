```python
import emptyApertureAnalysis as eaa

# name of noise-normalized, object subtracted image
nn = "macs0416par_f160w_noise-normalized.fits.gz"

# default list of apertures in arcsec
apertures = [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5,
             0.55, 0.6,0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95,
             1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.75, 2.0]

ref_aper = 0.7

sigmas, alpha, beta = eaa.run(
                                imname=nn,
                                outfile="macs0416par_f160w",
                                pixscl=0.06,
                                apertures=apertures,
                                ref_aper=ref_aper,
                                naper=1000,
                                figures=True,
                                title='M0416-par F160W',
                                nbins=35,
                                readfromfile=False
                             )
```

```python
import numpy as np

ref_sigma = sigmas[np.where(np.array(apertures) == ref_aper)[0][0]]

print(alpha, beta, ref_sigma)
```
